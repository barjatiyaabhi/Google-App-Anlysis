```python
df['Rating'] = df['Rating'].fillna(df['Rating'].median())# filling the data with fillna using median because the data is right skwed
```

```python
df.isnull().sum()# Sanity check
```

```python
df['Type'] = df['Type'].fillna(df['Type'].mode().values[0])#Filling the missing values 
```

```python
df['Current Ver'] = df['Current Ver'].fillna(df['Current Ver'].mode().values[0])
```

```python
df['Android Ver'] = df['Android Ver'].fillna(df['Android Ver'].mode().values[0])
```

```python
df.isnull().sum()#Sanity check
```

```python
df.info()
```

```python
df.head()# again checking the first five records
```

```python
df["Reviews"] = df["Reviews"].astype("int64")# Changing the datatype 
```

```python
df['Installs'] = df['Installs'].apply(lambda x: str(x).replace('+', '') if '+' in str(x) else str(x))#using apply and lamda and replacing the + with '' 
df['Installs'] = df['Installs'].apply(lambda x: str(x).replace(',', '') if ',' in str(x) else str(x))#using apply and lamda and replacing the , with '' 
df['Installs'] = df['Installs'].apply(lambda x: float(x))#changing the datatype using apply into float
```

```python
df['Size'] = df['Size'].apply(lambda x: str(x).replace('M', '') if 'M' in str(x) else str(x))
df['Installs'] = df['Installs'].apply(lambda x: int(x))
```

## Data Visualization

```python
df.head()
```

```python
grp = df.groupby('Category')# using groupby in category variable
x = grp['Rating'].agg(np.mean)#grouping the rating variable with aggregate of mean of rating
y = grp['Price'].agg(np.sum)#grouping the price variable with aggregate of sum
z = grp['Reviews'].agg(np.mean)##grouping the reviews variable with aggregate of mean of reviews
print(x)#using hard prints to print x,y,z
print(y)
print(z)
```

```python
plt.figure(figsize=(12,5))# plotting the scatter plot using the x variable i.e.mean of Raing vriable
plt.plot(x, "ro", color='g')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(16,5))# Category wise rating using the scatter plot
plt.plot(x,'ro', color='r')
plt.xticks(rotation=90)
plt.title('Category wise Rating')
plt.xlabel('Categories-->')
plt.ylabel('Rating-->')
plt.show()
```

```python
plt.figure(figsize=(16,5))# Category wise Reviews using the scatter plot
plt.plot(z,'bs', color='g')
plt.xticks(rotation=90)
plt.title('Category wise Reviews')
plt.xlabel('Categories-->')
plt.ylabel('Reviews-->')
plt.show()
```

### Ratings of Education and Events are very high as compared to the other categories
### Dating app has the lowest rating
### Communication and Social app has the highest number of reviews
